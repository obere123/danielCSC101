```python
print("This is a Program to calculate roots of a quadratic equation :ax^2+bx+c")
a=float(input("a=  "))
b=float(input("b=   "))
c=float(input("c=  "))
d=b**2-4*a*c
if d>0:
    ans1=(-b+(d**0.5))/(2*a)
    ans2=(-b-(d**0.5))/(2*a)
    print("The first solution is ",ans1)
    print("The second solution is",ans2)
elif d==0:
    ans3=-b/(2*a)
    print("The only solution is",ans3)
else:
    print("Imaginary roots")
```
